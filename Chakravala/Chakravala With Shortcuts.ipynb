```python
def Chakravala(n):
    
      
    n=int(n)
    p=1
    while p+1<sqrt(n):
      p=p+1
    q=1
    m=(p**2)-(n)*(q**2)
    m=int(m)

    print(f"\nInitial approximation")
    print(f"p={p}")
    print(f"q={q}")
    print(f"m={m}\n")
    #print(f"(p**2)-(n*(q**2))={(p**2)-(n*(q**2))}\n")

    #Shortcuts----------------------------------------------------------------------------------------------------------------------------------
    if m==-1:
      old_p=p
      old_q=q
      
      p=2*((old_p)**2)+1
      q=2*old_p*old_q
      m=(p**2)-(n*(q**2))
        
      print(f"Shortcut implemented")
      print(f"p={p}")
      print(f"q={q}")
      print(f"m={(p**2)-(n*(q**2))}\n")
        
    elif m==2:
      old_p=p
      old_q=q
      
      p=((old_p)**2)-1
      q=old_p*old_q
      m=(p**2)-(n*(q**2))
      
      print(f"Shortcut implemented")
      print(f"p={p}")
      print(f"q={q}")
      print(f"m={(p**2)-(n*(q**2))}\n")
      
    elif m==-2:
      old_p=p
      old_q=q
      
      p=((old_p)**2)+1
      q=old_p*old_q
      m=(p**2)-(n*(q**2))
      
      print(f"Shortcut implemented")
      print(f"p={p}")
      print(f"q={q}")
      print(f"m={(p**2)-(n*(q**2))}\n")  
      
    elif m==4:
      if p%2==0:
        old_p=p
        old_q=q
      
        p=((old_p**2)-2)//2
        q=(old_p*old_q)//2
        m=(p**2)-(n*(q**2))
      
        print(f"Shortcut implemented")
        print(f"p={p}")
        print(f"q={q}")
        print(f"m={(p**2)-(n*(q**2))}\n")
      
      else:
        old_p=p
        old_q=q
       
        p=(old_p*((old_p**2)-3))//2
        q=(old_q*((old_p**2)-1))//2
        m=(p**2)-(n*(q**2))
       
        print(f"Shortcut implemented")
        print(f"p={p}")
        print(f"q={q}")
        print(f"m={(p**2)-(n*(q**2))}\n")
      
    elif m==-4:
      
      if p%2==0:
        old_p=p
        old_q=q

        p=((old_p**2)+2)//2
        q=(old_p*old_q)//2
        m=(p**2)-(n*(q**2))

        print(f"Shortcut implemented")
        print(f"p={p}")
        print(f"q={q}")
        print(f"m={(p**2)-(n*(q**2))}\n")
      
      else:
        old_p=p
        old_q=q

        p=(((old_p**2)+2)*((((old_p**2)+1)*((old_p**2)+3))-2))//2
        q=(old_p*old_q*((old_p**2)+1)*((old_p**2)+3))//2
        m=(p**2)-(n*(q**2))
         
        print(f"Shortcut implemented")
        print(f"p={p}")
        print(f"q={q}")
        print(f"m={(p**2)-(n*(q**2))}\n")  
    
    
    # First iteration----------------------------------------------------------------------------------------------------------------------------------
    if m!=1:
      x=abs(m)-p
      x=int(x)
      j=2
      while x+abs(m)<sqrt(n):
        x=abs(m)*j-p
        x=int(x)
        j=j+1
      
     
      old_p=p
      old_q=q
      old_m=m

      p=((old_p*x)+n)//abs(old_m)
      q=(old_p+x)//abs(old_m)
      m=(x**2-(n))//old_m

      p=int(p)
      q=int(q)
      m=int(m)                       
      
      print("Iteration 1 ")
      print(f"x={x}")
      print(f"p={p}")
      print(f"q={q}")
      print(f"m={m}\n") 
      #print(f"(p**2)-(n*(q**2))={(p**2)-(n*(q**2))}\n") 
        
    #Shortcuts------------------------------------------------------------------------------------------------------------------------------------------

    if m==-1:
      old_p=p
      old_q=q
      
      p=2*((old_p)**2)+1
      q=2*old_p*old_q
      m=(p**2)-(n*(q**2))
        
      print(f"Shortcut implemented")
      print(f"p={p}")
      print(f"q={q}")
      print(f"m={(p**2)-(n*(q**2))}\n")
        
    elif m==2:
      old_p=p
      old_q=q
      
      p=((old_p)**2)-1
      q=old_p*old_q
      m=(p**2)-(n*(q**2))
      
      print(f"Shortcut implemented")
      print(f"p={p}")
      print(f"q={q}")
      print(f"m={(p**2)-(n*(q**2))}\n")
      
    elif m==-2:
      old_p=p
      old_q=q
      
      p=((old_p)**2)+1
      q=old_p*old_q
      m=(p**2)-(n*(q**2))
      
      print(f"Shortcut implemented")
      print(f"p={p}")
      print(f"q={q}")
      print(f"m={(p**2)-(n*(q**2))}\n")  
      
    elif m==4:
      if p%2==0:
        old_p=p
        old_q=q
      
        p=((old_p**2)-2)//2
        q=(old_p*old_q)//2
        m=(p**2)-(n*(q**2))
      
        print(f"p={p}")
        print(f"q={q}")
        print(f"m={(p**2)-(n*(q**2))}\n")
      
      else:
        old_p=p
        old_q=q
       
        p=(old_p*((old_p**2)-3))//2
        q=(old_q*((old_p**2)-1))//2
        m=(p**2)-(n*(q**2))
       
        print(f"Shortcut implemented")
        print(f"p={p}")
        print(f"q={q}")
        print(f"m={(p**2)-(n*(q**2))}\n")
      
    elif m==-4:
      
      if p%2==0:
        old_p=p
        old_q=q

        p=((old_p**2)+2)//2
        q=(old_p*old_q)//2
        m=(p**2)-(n*(q**2))

        print(f"Shortcut implemented")
        print(f"p={p}")
        print(f"q={q}")
        print(f"m={(p**2)-(n*(q**2))}\n")
      
      else:
        old_p=p
        old_q=q

        p=(((old_p**2)+2)*((((old_p**2)+1)*((old_p**2)+3))-2))//2
        q=(old_p*old_q*((old_p**2)+1)*((old_p**2)+3))//2
        m=(p**2)-(n*(q**2))
         
        print(f"Shortcut implemented")
        print(f"p={p}")
        print(f"q={q}")
        print(f"m={(p**2)-(n*(q**2))}\n")


    
    # Consecutive iterations------------------------------------------------------------------------------------------------------------------------
    i=1
    while m!=1:
      #print(f"x+abs(m)={x+abs(m)}, sqrt(n)={sqrt(n)}")
      old_x=x
      x=abs(m)-old_x
      x=int(x)
      j=2
      while x+abs(m)<sqrt(n):
        x=abs(m)*(j)-old_x
        x=int(x)
        j=j+1
      
      old_p=p
      old_q=q
      old_m=m

      p=int(int(old_p*x)+int(n*old_q))//abs(old_m)
      q=int(old_p+int(x*old_q))//abs(old_m)
      m=(x**2-(n))//old_m 

      p=int(p)
      q=int(q)
      m=int(m)  

      print(f"Iteration {i+1}")
      print(f"x={x}")  
      print(f"p={p}")
      print(f"q={q}")
      print(f"m={m}\n")
      #print(f"(p**2)-(n*(q**2))={int(p**2)-(n*int(q**2))}\n")

      #Shortcuts----------------------------------------------------------------
      if m==-1:
        old_p=p
        old_q=q
      
        p=2*((old_p)**2)+1
        q=2*old_p*old_q
        m=(p**2)-(n*(q**2))
      
        print(f"Shortcut implemented")
        print(f"p={p}")
        print(f"q={q}")
        print(f"m={(p**2)-(n*(q**2))}\n")
      
      elif m==2:
        old_p=p
        old_q=q
      
        p=((old_p)**2)-1
        q=old_p*old_q
        m=(p**2)-(n*(q**2))
      
        print(f"Shortcut implemented")
        print(f"p={p}")
        print(f"q={q}")
        print(f"m={(p**2)-(n*(q**2))}\n")
      
      elif m==-2:
        old_p=p
        old_q=q
      
        p=((old_p)**2)+1
        q=old_p*old_q
        m=(p**2)-(n*(q**2))
      
        print(f"Shortcut implemented")
        print(f"p={p}")
        print(f"q={q}")
        print(f"m={(p**2)-(n*(q**2))}\n")  
      
      elif m==4:
        if p%2==0:
          old_p=p
          old_q=q
                   
          p=((old_p**2)-2)//2
          q=(old_p*old_q)//2
          m=(p**2)-(n*(q**2))
         
          print(f"Shortcut implemented")
          print(f"p={p}")
          print(f"q={q}")
          print(f"m={(p**2)-(n*(q**2))}\n")
        
        else:
          old_p=p
          old_q=q
            
          p=(old_p*((old_p**2)-3))//2
          q=(old_q*((old_p**2)-1))//2
          m=(p**2)-(n*(q**2))
           
          print(f"Shortcut implemented")
          print(f"p={p}")
          print(f"q={q}")
          print(f"m={(p**2)-(n*(q**2))}\n")
       
      elif m==-4:
             
        if p%2==0:
          old_p=p
          old_q=q
               
          p=((old_p**2)+2)//2
          q=(old_p*old_q)//2
          m=(p**2)-(n*(q**2))
                 
          print(f"Shortcut implemented")
          print(f"p={p}")
          print(f"q={q}")
          print(f"m={(p**2)-(n*(q**2))}\n")
             
        else:
          old_p=p
          old_q=q
                
          p=(((old_p**2)+2)*((((old_p**2)+1)*((old_p**2)+3))-2))//2
          q=(old_p*old_q*((old_p**2)+1)*((old_p**2)+3))//2
          m=(p**2)-(n*(q**2))
                    
          print(f"Shortcut implemented")
          print(f"p={p}")
          print(f"q={q}")
          print(f"m={(p**2)-(n*(q**2))}\n")
        
        
            
      i=i+1
      
    return p,q,m


```

```python
def samasabhavana(n,p,q,m):
    
    n=int(n)
    sol_list=[(p,q)]
    
    if m>1:  
      
      old_p=p
      old_q=q
    
      p=old_p**2+(n*(old_q**2))
      q=(old_p*old_q)+(old_q*old_p)
      sol_list.append((p,q))
      
      while len(sol_list)!=m:
        x1=choice(sol_list)
        x2=choice(sol_list)

        #print(x1)
        p1=x1[0] 
        q1=x1[1]
        p2=x2[0]
        q2=x2[1]

        p3=(p1*p2)+(n*q1*q2)
        q3=(p1*q2)+(q1*p2)
        if (p3,q3) not in sol_list:
          sol_list.append((p3,q3))

    return sol_list  
      
```

```python
def main():
    #print("Enter an non-perfect square integer >1 :")
    n=float(input("Enter an non-perfect square integer >1 :"))
    if n>1 and int(sqrt(n))!=sqrt(n) and int(n)==n:
      
      p,q,m=Chakravala(n)
      n=int(n)
      #print(f"\n{p}, {q}")
      #print(f"\nCheck output {p}**2-({n}*({q}**2))={p**2-(n*(q**2))}")
      m=int(input("\nEnter the number of solutions needed:"))
      if m<1:
        print("Invalid Input")
      else:  
        print(f"\n")
        sol_list=samasabhavana(n,p,q,m)

        i=1
        for x in sol_list:
          print(f"{i}. {x}")
          p,q=x
          print(f"   Check output {p}**2-({n}*({q}**2))={p**2-(n*(q**2))}\n")
          i=i+1    
    
    elif int(n)!=n and n<=1:
      print("\nInvalid input. The input is not an integer and is <=1.")
    elif int(sqrt(n))==sqrt(n) and n<=1:
      print("\nInvalid input. The input is a perfect square and is<=1.")
    elif int(sqrt(n))==sqrt(n):
      print("\nInvalid input. The input is a perfect square.")
    elif int(n)!=n:
      print("\nInvalid input. The input is not an integer.")
    elif n<=1:
      print("\nInvalid input. The input is <=1.") 

  



             
       
```
